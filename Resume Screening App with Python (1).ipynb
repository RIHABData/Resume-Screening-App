## Cleaning Data

```python
Resume['Category'] = Resume['Category'].str.strip().str.lower()
```

## Vactorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(Resume['Resume_str'])
requredTaxt = tfidf.transform(Resume['Resume_str'])
```

```python
requredTaxt
```

```python
Resume
```

## Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requredTaxt, Resume['Category'], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Now let's train the model and print the classification report

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

# Create the classifier instance
clf = OneVsRestClassifier(KNeighborsClassifier(n_neighbors=5))

# Fit the model
clf.fit(X_train, y_train)

# Predict
ypred = clf.predict(X_test)

# Evaluate
print("Predictions:", ypred)
print("Accuracy:", accuracy_score(y_test, ypred))
```
